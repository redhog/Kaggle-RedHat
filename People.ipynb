```python
people = np.load("people.npz")['x']
```

```python
fig, ax = plt.subplots()
N, bins, patches = ax.hist(people['date'], 100)
plt.show()
```

```python
for name in people.dtype.names:
    if not name.startswith("char_"): continue
    fig, ax = plt.subplots()
    N, bins, patches = ax.hist(people[people[name] == 0.0]['date'], 100, color="r")
    N, bins, patches = ax.hist(people[people[name] == 1.0]['date'], 100, color="b")
    ax.set_title(name)
    plt.show()
```
